# Visualization 

```python
sns.pairplot(df, hue = 'sepal length', size = 3.5, aspect = 1)
```

```python
sns.pairplot(df, hue = 'class', size = 3, aspect = 1)
```

```python
df.hist(edgecolor = 'yellow', figsize = (12, 9), linewidth = 1.5)
```

```python
plt.Figure(figsize = (13, 9))
sns.violinplot(x = 'petal width', y = 'petal length', data = df, gridsize = 150, width = 0.9)
plt.show()
```

```python
plt.figure(figsize = (12, 10))
plt.subplot(2, 2, 1)
sns.violinplot(x = 'petal length', y = 'class', data = df)
plt.subplot(2, 2, 2)
sns.violinplot(x = 'petal width', y = 'class', data = df)
plt.subplot(2, 2, 3)
sns.violinplot(x = 'sepal width', y = 'class', data = df)
plt.subplot(2, 2, 4)
sns.violinplot(x = 'sepal length', y = 'class', data = df)
plt.show()
```
